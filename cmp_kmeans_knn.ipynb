```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split



data = load_iris()
X , y = data.data[:,2:4], data.target
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.33, random_state=42)
```

## KMeans

```python
kmeans = KMeans(n_clusters=3,n_init=10)
kmeans.fit(Xtr)

for i in range(len(set(yte))):
  plt.plot(Xte[yte==i, 0],
           Xte[yte==i, 1], 's'
           , label=data.target_names[i])

ypred_kmeans = kmeans.predict(Xte)
for i in range(len(set(ypred_kmeans))):
  plt.plot(Xte[ypred_kmeans==i, 0],
           Xte[ypred_kmeans==i, 1], '.'
           , label=data.target_names[i])
accuracy_kmeans = accuracy_score(yte, ypred_kmeans)

plt.xlabel('Petal Length')
plt.ylabel('Petal width')
plt.legend()
plt.title('Figura 1')
plt.show()
accuracy_kmeans
```

## KNeighbors

```python
knn = KNeighborsClassifier()
knn.fit(Xtr,ytr)
for i in range(len(set(yte))):
  plt.plot(Xte[yte==i, 0],
           Xte[yte==i, 1], 's'
           , label=data.target_names[i])

ypred_knn = knn.predict(Xte)
for i in range(len(set(ypred_knn))):
  plt.plot(Xte[ypred_knn==i, 0],
           Xte[ypred_knn==i, 1], '.'
           , label=data.target_names[i])
accuracy_knn = accuracy_score(yte, ypred_knn)

plt.xlabel('Petal Length')
plt.ylabel('Petal width')
plt.legend()
plt.title('Figura 2')
plt.show()
accuracy_knn
```

Embora o KMeans tenha conseguido 0.38 de acurácia ele conseguiu classificar grande parte dos dados. O que acarretou esse valor, foi a troca da categoria da iris, conforme apresentado na Figura 1, o que era *versicolor* foi trocado por *virginica* e vice-versa. Como o KNN já conhecia as categorias ele apresentou uma acurácia de 0.98. Portanto, o KNN apresentou um desempenho melhor comparado ao KMeans.
