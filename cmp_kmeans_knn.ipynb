```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split



data = load_iris()
X , y = data.data[:,2:4], data.target
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.33, random_state=42)
```

## KMeans

```python
kmeans = KMeans(n_clusters=3,n_init=10)
kmeans.fit(X)

for i in range(len(set(y))):
  plt.plot(X[y==i, 0],
           X[y==i, 1], 's'
           , label=data.target_names[i])

ypred_kmeans = kmeans.predict(X)
for i in range(len(set(ypred_kmeans))):
  plt.plot(X[ypred_kmeans==i, 0],
           X[ypred_kmeans==i, 1], '.'
           , label=data.target_names[i])
accuracy_kmeans = accuracy_score(y, ypred_kmeans)

plt.xlabel('Petal Length')
plt.ylabel('Petal width')
plt.legend()
plt.show()
accuracy_kmeans
```

## KNeighbors

```python
knn = KNeighborsClassifier()
knn.fit(X,y)
for i in range(len(set(y))):
  plt.plot(X[y==i, 0],
           X[y==i, 1], 's'
           , label=data.target_names[i])

ypred_knn = knn.predict(X)
for i in range(len(set(ypred_knn))):
  plt.plot(X[ypred_knn==i, 0],
           X[ypred_knn==i, 1], '.'
           , label=data.target_names[i])
accuracy_knn = accuracy_score(y, ypred_knn)

plt.xlabel('Petal Length')
plt.ylabel('Petal width')
plt.legend()
plt.show()
accuracy_knn
```
